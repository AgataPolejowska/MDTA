```python
import pandas as pd
import sklearn.cluster as cluster


data = pd.read_csv("dataset.csv")

kmeans = cluster.KMeans(n_clusters=3, init="k-means++")
kmeans = kmeans.fit(data,
                                [['age_bins',
                                  'gender',
                                  'place',
                                  'wspomaganie_innowacji_Tak',
                                  'ai_zastapienie_lekarzy_Tak',
                                  'operacja_robot_chirurg_Jedynie przez wykształconego chirurga',
                                  'ingerencja_geny_Tak',
                                  'dane_czujniki_spersonalizowanie_Tak',
                                  'leczenie_leki_genetyczne_Tak',
                                  'opiekun_robot_czlowiek_Człowiek',
                                  'koszulka_smart_Tak',
                                  'tatuaz_smart_Tak',
                                  'chip_Tak',
                                  'Zapłacić więcej za nowsze, bardziej precyzyjne technologie',
                                  'stosowanie_urzadzen_Tak',
                                  'udzial_zaawansowane_badanie_Nie wiem',
                                  'udzial_zaawansowane_badanie_Tak',
                                  'vr_Tak',
                                  'predyspozycje_genetyczne_Tak',
                                  'liczba_daVinci_2 sztuki',
                                  'liczba_daVinci_około 10 sztuk',
                                  'liczba_daVinci_około 100 sztuk',
                                  'liczba_daVinci_około 1000 sztuk',
                                  'przewidywanie_stan_zdrowia_Tak',
                                  'swiadomosc_korzysci_ai_Tak',
                                  'telemedycyna_Jako uzupełnienie tradycyjnych form opieki medycznej',
                                  'telemedycyna_Tylko ze względu na epidemię COVID-19',
                                  'telemedycyna_Zamiast tradycyjnej opieki medycznej']])
```

```python
data['Gotowosc'] = kmeans.labels_
data.head()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

iris = load_iris()
X = pd.DataFrame(data, columns=data['Gotowosc'])
#print(X)
data = X[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']]

sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(data)
    data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```
