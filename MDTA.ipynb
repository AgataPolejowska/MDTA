#### One Hot Encoding

***
### Tabela FEARS

```python
pd.set_option('display.max_rows', None)

wspomaganie_innowacji_dummies = pd.get_dummies(df_fears.wspomaganie_innowacji, prefix="wspomaganie_innowacji")
wspomaganie_innowacji_dummies

df_fears = pd.concat([df_fears, wspomaganie_innowacji_dummies], axis='columns')

df_fears
```

```python
# Usuniecie niepotrzebnych kolumn
df_fears = df_fears.drop(['wspomaganie_innowacji_Nie'], axis='columns')
df_fears
```

```python
pd.set_option('display.max_rows', None)

operacja_robot_chirurg_dummies = pd.get_dummies(df_fears.operacja_robot_chirurg)
operacja_robot_chirurg_dummies

df_fears = pd.concat([df_fears, operacja_robot_chirurg_dummies], axis='columns')

df_fears
```

```python
# Usuniecie niepotrzebnych kolumn
df_fears = df_fears.drop(['operacja_robot_chirurg', 'Tylko przez robota medycznego (w przypadku 99% pewności jego prawidłowego działania)'], axis='columns')
df_fears
```

```python
pd.set_option('display.max_rows', None)

ai_zastapienie_lekarzy_dummies = pd.get_dummies(df_fears.ai_zastapienie_lekarzy, prefix="ai_zastapienie_lekarzy")
ai_zastapienie_lekarzy_dummies

df_fears = pd.concat([df_fears, ai_zastapienie_lekarzy_dummies], axis='columns')

df_fears
```

```python
# Usuniecie niepotrzebnych kolumn
df_fears = df_fears.drop(['ai_zastapienie_lekarzy', 'ai_zastapienie_lekarzy_Nie'], axis='columns')
df_fears
```

```python
df_fears = df_fears.drop(['wspomaganie_innowacji'], axis='columns')
df_fears
```

```python
df_fears.to_csv('fears_processed.csv')
```
